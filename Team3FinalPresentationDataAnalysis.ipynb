```python
import pandas as pd
import numpy as np
import plotly.express as px
# import 
```

```python
#Read Original data
Original_data = pd.read_excel('WDIW Dataset.xlsx',index_col =0)
Original_data
```

```python
#Create Dataset for South East Africa
South_East_Africa = Original_data[Original_data['Cool Name'] == 'Jazz']
South_East_Africa
```

```python
#Drop 3 countries, Somalia, South Sudan, Eritrea
South_East_Africa_19 = South_East_Africa.drop(['SSD','SOM','ERI'])
South_East_Africa_19
```

```python
# Check for Missing Value(NaN)
South_East_Africa_19.isnull()
```

```python
# Count All Missing Values in Suoth East Africa
print("All Missing Value :", South_East_Africa.isnull().sum().sum())
```

```python
# Count All Missing Values in 19 Countries
print("All Missing Value :", South_East_Africa_19.isnull().sum().sum())
```

```python
#Count Missing Values Per Column
#print("Missing Value per Column :\n", South_East_Africa.isnull().sum())
MissingValue = South_East_Africa_19.isnull().sum()

South_East_Africa_19_wMissing = South_East_Africa_19.append(MissingValue,ignore_index=True)
South_East_Africa_19_wMissing
```

```python
# Count Missing Value Per Row
for i in range(len(South_East_Africa_19.index)) :
        print("Nan in", South_East_Africa_19.iloc[[i],[0]] , " : " ,  South_East_Africa_19.iloc[i].isnull().sum())
```
